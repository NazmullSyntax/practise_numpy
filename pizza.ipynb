```python
import numpy as np

# Pizza names
pizza_names = np.array([
    "Margherita",
    "Pepperoni",
    "BBQ Chicken",
    "Veggie",
    "Cheese"
])

# Price of each pizza
pizza_prices = np.array([350, 450, 500, 400, 380])

# Quantity sold
pizza_sold = np.array([20, 15, 10, 12, 18])

# Revenue calculation
revenue = pizza_prices * pizza_sold

# Total revenue
total_revenue = np.sum(revenue)

# Total pizzas sold
total_pizzas = np.sum(pizza_sold)

# Average pizzas sold
average_sales = np.mean(pizza_sold)

# Highest selling pizza
max_index = np.argmax(pizza_sold)

print("===== PIZZA SHOP REPORT =====")
print("Pizza Names:", pizza_names)

print("\nRevenue for each pizza:")
for i in range(len(pizza_names)):
    print(f"{pizza_names[i]} = {revenue[i]} Tk")

print("\nTotal Revenue:", total_revenue, "Tk")
print("Total Pizzas Sold:", total_pizzas)
print("Average Sales:", average_sales)

print("\nBest Selling Pizza:")
print(pizza_names[max_index], "-", pizza_sold[max_index], "pieces sold")
```
